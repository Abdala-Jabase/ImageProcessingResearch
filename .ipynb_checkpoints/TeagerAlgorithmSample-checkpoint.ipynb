# Teager Operator

## Part 1: Teager of a Signal

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 30, 1000)

y = np.cos((2/10)*x+10)+np.cos((4/10)*x+2)+np.sin(x+3)+np.sin(3*x+2)+3.5

plt.plot(x, y)
plt.xlabel('Time (s):')
plt.ylabel('Voltage (V):')
plt.show()
```

### Above is the signal incoming from a particular sensor

#### The signal we be discreted at a sampling period of 1 seconds

```python
samplingPeriod = 1
samples = int(30/samplingPeriod)
discretex = np.linspace(0 ,30 , samples)

y = np.cos((2/10)*discretex+10)+np.cos((4/10)*discretex+2)+np.sin(discretex+3)+np.sin(3*discretex+2)+3.5

plt.plot(discretex, y, '.r')
plt.xlabel('Time (s):')
plt.ylabel('Voltage (V):')
plt.show()
```
